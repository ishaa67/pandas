```python
#6. AND (&)
#1. Display records where Sales > 500 and Profit > 100.
import pandas as pd 
data = pd.read_excel("pandas.xlsx")


data[(data['Sales'] > 500) & (data['Profit'] > 100)]
```

```python
#2. Display customers from the West region and Furniture category.
data[(data["Region"] == "West") & (data["Product Category"] =="Furniture")]
```

```python
#3. Display records where Discount > 0.05 and Shipping Cost > 20.
data[(data["Discount"]>0.05) & (data["Shipping Cost"])]
```

```python
#4. Display orders where Quantity ordered new > 10 and Profit > 50.
data[(data["Quantity ordered new"]>10) & (data["Profit"]>50)]
```

```python
#5. Display products where Product Category = Office Supplies and Sales > 1000.
data[(data["Product Category"]=="Office Supplies")&(data["Sales"]>1000)]
```

```python
#6. Display records where Region = Central and Profit < 0.
data[(data["Region"]=="Central")&(data["Profit"]>0)]
```

```python
#7. Display orders where Ship Mode = Regular Air and Sales > 200.
data[(data["Ship Mode"]=="Regular Air")&(data["Sales"]>200)]
```

```python
#7. OR (|)
#1. Display records where Region = East or Region = West.
data[(data["Region"]=="East")|(data["Region"]=="West")]
```

```python
#2. Display products where Product Category = Furniture or Technology.
data[(data["Product Category"]=="Furniture")|(data["Product Category"]=="Technology")]
```

```python
#3. Display records where Sales > 1000 or Profit > 500.
data[(data["Sales"]>1000)|(data["Profit"]>500)]
```

```python
#4. Display orders where Discount > 0.10 or Shipping Cost > 30.
data[(data["Discount"]>0.10)|(data["Shipping Cost"]>30)]
```

```python
#5. Display customers from California or Texas.
data[(data["State or Province"]=="California")|(data["State or Province"]=="Texas")]
```

```python
#6. Display orders where Ship Mode = Express Air or Delivery Truck.
data[(data["Ship Mode"]=="Express Air")|(data["Ship Mode"]=="Delivery Truck")]
```

```python
#7. Display records where Profit < 0 or Sales > 2000.
data[(data["Profit"]<0)|(data["Sales"]>2000)]
```
